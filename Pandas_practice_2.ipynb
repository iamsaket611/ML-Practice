```python
#12. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

transac.join(users,rsuffix='users').head(5)
```

```python
pd.merge(transac,users,how='left')
```

```python
#13. Which transactions have a UserID not in users?
pd.merge(transac,users,how='inner',on='UserID')
```

```python
#15. Join users to transactions, displaying all matching rows AND all non-matching rows (full
pd.merge(transac,users,how='outer')
```

```python
#16. Determine which sessions occurred on the same day each user registered
pd.merge(users,transac,right_on='TransactionDate',left_on='Registered')
```
