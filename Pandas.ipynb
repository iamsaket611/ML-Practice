```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/PramodShenoy/911-Calls/master/911.csv")
df.head()
```

```python
# 1.Describe for Numerical and Categorical Values

## for Numerical Values
df.describe()
## for categorical values
df_C = df.dtypes[df.dtypes=='object'].index
df[df_C].describe()
```

```python
# 2.) TypeCast Latitude into 'INT' and then then perform groupby operation on it and the count on it
df['lat1'] = df['lat'].astype(int)
grp = df.groupby('lat1')
df2= pd.DataFrame(grp)
df2[0].count()
```

```python
# 3.) how many people are from same zip locatio
df[['zip']].apply(pd.value_counts)
df['zip'].value_counts()
```

```python
## 4.check time stamp datatype and extract day in new col
time_stamps = pd.to_datetime(df['timeStamp'])
type(time_stamps)
df['day'] = time_stamps.dt.day
df
```

```python
##5. filter out all the zip codes which address starts with 'A'
y =[]
[y.append(df['zip'][i])  for i in range(len(df)) if str(df['addr'][i])[0]== 'A']

#OR
x = []
for i in range(len(df)):
    if (str(df['addr'][i])[0]) == 'A':
        x.append(df['zip'][i])
df['new_zip'] = pd.DataFrame(x)
df.head()        
```

```python
## 6) try to map timestamp with day like monday, tuesday....
df_ts = pd.to_datetime(df['timeStamp'])
df['days_ts'] = df_ts.dt.day_name()
df.tail(10)
```

```python
#7) from title extract service detail
df_tt = []
df_ftt = []
tt = df['title']
for i in tt:
    df_tt.append(i.split(sep=(':')))
for j in df_tt:
    df_ftt.append(j[1])
df['tile_new'] = pd.DataFrame(df_ftt)
df.head(10)
```
