```python
df3 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
df3.head()
```

```python
#Convert the type of the column Year to datetime64
```

```python
df3['y'] = pd.to_datetime(df3['Year'],format='%Y')
df3.head()
```

```python
#Set the Year column as the index of the dataframeÂ¶
df3.set_index(df3['y'],drop =True,inplace =True)
df3.head()
```

```python
#Uses resample to sum each decade
crimes = df3.resample('10AS').sum()
crimes
```

```python
# Uses resample to get the max value only for the "Population" column
population = crimes['Population'].resample('10AS').max()
Population
```

```python
# Updating the "Population" column
crimes['Population'] = Population
crimes
```
